## **Reading the Data from JSON file**


```python
import json
fhand = open("Menu.json",'r')
r = fhand.read()
fhand.close()

List = json.loads(r)
```

```python
List
```

## **Adding New Items to Inventory**

```python
while True:
  prod_id = input("Enter Product ID You Want to Add : ")
  if prod_id == 'done':
    break
  elif prod_id in List:
    quant = int(input("Enter the Quantity : "))
    List[prod_id]["quant"] += quant
    jsn = json.dumps(List)
    fhand = open("Menu(Updated).json",'a')
    fhand.write(jsn)
    fhand.close()
    continue
  else:
    ui_name = input("Enter The Name of the Product : ")
    ui_quant = int(input("Enter the Quantity of the Product : "))
    ui_price = int(input("Enter the Price of the Product :"))
    ui_discount = int(input("Enter the Discount(in percentage) : "))
    ui_category = input("Enter the Category of the Product : ")

    List[prod_id] = {'name': ui_name, 'price': ui_price, 'quant': ui_quant, 'discount(in percentage)': ui_discount,'category': ui_category}

    jsn = json.dumps(List)

    fhand = open("Menu(Updated).json",'a')
    fhand.write(jsn)
    fhand.close()
```
