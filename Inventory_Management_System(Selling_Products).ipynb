## **Inventory Reading**

```python
import json

fhand = open("Record.json","r")
r = fhand.read()
fhand.close()

records = json.loads(r)
```

## **User Input and JSON Sales**

```python
print("Hello Sir, Welcome to Our Supermarket")
print()
print("-------------------------------------")
print()
count = 0
while True:
  ui_prod  = str(input("Enter The Product ID That You Want to Purchase : "))
  print()
  if ui_prod == "done":
    break
  elif ui_prod not in records:
    print("Sorry sir we dont have the product")
    print()
    continue
  else:  
    ui_quant = int(input("Enter The Quantity : "))
    print()
    print('*************************************')
    print()
    print("Product Purchased : ", records[ui_prod]['name'])
    print("Price Of Product : ", records[ui_prod]['price'])
    print("Category of Product : ",records[ui_prod]['category'])
    print("Discount Availed : ",records[ui_prod]['discount(in percentage)'],"%")
    print("Amount To Be Paid: ", (records[ui_prod]['price'] * ui_quant)-(records[ui_prod]['price'] *records[ui_prod]['discount(in percentage)']/100))
    print()
    print('*************************************')
    print()
    records[ui_prod]['quant'] = records[ui_prod]['quant'] - ui_quant
    
 #JSON Sales   
    count += 1
    r = {str(count):{'Prod_ID':ui_prod,'name':records[ui_prod]['name'] , 'quant' : ui_quant, 'price':(records[ui_prod]['price'] * ui_quant)-(records[ui_prod]['price'] *records[ui_prod]['discount(in percentage)']/100) ,'category':records[ui_prod]['category']}}
    res = json.dumps(r)
    res = json.dumps(',')
    fd = open("Sales.json","a")
    fd.write(res)
    fd.close()
    


print()
print("Thank You! Sir Visit again")
```
